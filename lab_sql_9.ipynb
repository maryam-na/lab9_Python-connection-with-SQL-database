```python
# importing libraries
import pymysql
from sqlalchemy import create_engine
import pandas as pd
import getpass
password = getpass.getpass()
```

```python
# Create a Python connection with SQL database
connection_string = 'mysql+pymysql://root:'+password+'@localhost/bank'
engine = create_engine(connection_string)

# Create a Python connection with SQL database and retrieve the results of the last two queries (also mentioned below) as dataframes:
# creating a dataframe for May
query = "SELECT * FROM sakila.rental  WHERE  MONTH (rental_date) LIKE '%%5'"
rentals_may = pd.read_sql_query(query,engine)
rentals_may.head(20)
```

```python
# Creating a dataframe for June
query = "SELECT * FROM sakila.rental  WHERE  MONTH (rental_date) LIKE '%%6'"
rentals_june = pd.read_sql_query(query,engine)
rentals_june.head(20)
```

```python
# Creating a dataframe for customer_id and customer_name
query = """SELECT 
    customer_id, CONCAT((first_name), ' ', (last_name)) AS customer_name FROM sakila.customer;"""
customer = pd.read_sql_query(query,engine)
customer.head(20)
```

```python
# Check the number of rentals for each customer for May.
frequency_per_customer_may = pd.DataFrame(rentals_may.customer_id.value_counts())
frequency_per_customer_may.columns = ['frequency_for_may'] 
frequency_per_customer_may.index.names = ['customer_id']
frequency_per_customer_may
```

```python
# Check the number of rentals for each customer for June.
frequency_per_customer_june = pd.DataFrame(rentals_june.customer_id.value_counts())
frequency_per_customer_june.columns = ['frequency_for_june'] 
frequency_per_customer_june.index.names = ['customer_id']
frequency_per_customer_june
```

```python
# joining dataFrames of May and June
frequency = frequency_per_customer_may.merge(frequency_per_customer_june, how='inner', on='customer_id')
frequency
```

```python
# joining dataFrames frequency and customer to add customer_column to frequency
frequency = customer.merge(frequency, how='inner', on='customer_id')
frequency
```

```python
# Write a function that checks if customer borrowed more or less films in the month of June as compared to May.

def Number_borrowed_comparison(df):
    for index, row in df.iterrows():        
        if row['frequency_for_may'] < row['frequency_for_june']:
            print(f"{row['customer_name']}: More rentals in June")
        elif row['frequency_for_may'] > row['frequency_for_june']: 
            print(f"{row['customer_name']}:More rentals in May")
        else:
            print(f"{row['customer_name']}:Same rentals in May and June")

Number_borrowed_comparison(frequency)
```
